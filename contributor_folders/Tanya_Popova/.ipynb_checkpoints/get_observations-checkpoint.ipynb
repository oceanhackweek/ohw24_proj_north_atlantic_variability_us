```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.path as mpath
from matplotlib.colors import CenteredNorm
from scipy import stats, signal
from cartopy import crs as ccrs, feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
```

```python
data_url = 'https://psl.noaa.gov/thredds/dodsC/Datasets/noaa.ersst.v5/sst.mnmean.nc'
```

```python
sst = xr.open_dataset(data_url).sst
# sst = sst.sel(time=slice('2023','2024'))
sst
```

```python
sst.coords['lon'] = (sst.coords['lon'] + 180) % 360 - 180
sst = sst.sortby(sst.lon)
sst
```

```python
sst_slice = sst.sel(lat=slice(90,0), lon=slice(-99,13))
sst_slice
```

```python
sst_slice.isel(time=0).plot()
```

```python
import geoviews as gv
import geoviews.feature as gf
import xarray as xr
from cartopy import crs

gv.extension('bokeh', 'matplotlib')
```

```python
dataset = gv.Dataset(sst)
ensemble = dataset.to(gv.Image, ['lon', 'lat'], 'sst')

gv.output(ensemble.opts(cmap='viridis', colorbar=True, fig_size=120, backend='matplotlib') * gf.coastline(),
          backend='matplotlib')
```
